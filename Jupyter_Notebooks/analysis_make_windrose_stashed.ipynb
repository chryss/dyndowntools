```python
import plotly.express as px
from pathlib import Path
from windrose import WindroseAxes, plot_windrose
import pandas as pd
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
import seaborn as sns
import calendar as cal
```

```python
startdate = '1980-01-01'
enddate = '2019-12-31'
datadir = Path().absolute().parent / "evaluation/working"
stationdir = Path().absolute().parent / "evaluation/weatherstationdata/SNAP"
teststation = 'ANC_PANC'
windfp = datadir / f"wind_{teststation}_1980_2019_4km.csv"
filepattern_station = "alaska_airports_hourly_winds_{ioac}.csv"
stations = {
    "ANC_PANC": "PANC",
    "FAI_PAFA": "PAFA",
    "UTQ_PABR": "PABR",
    "BTH_PABE": "PABE"
}
locations = {
    "ANC_PANC": "Anchorage",
    "FAI_PAFA": "Fairbanks",
    "UTQ_PABR": "Utqiagvik",
    "BTH_PABE": "Bethel"   
}
windDF = pd.read_csv(windfp, index_col = 'Time', parse_dates = ['Time'])
# windDF = windDF.loc[windDF.wspd10 > 0.1]
windDF['wspd10'] = 2.23694 * windDF['wspd10']      # m/s to mph
windDF['month'] = windDF.index.month.map(dict(enumerate(cal.month_abbr)))    
windDF
```

```python
def get_dataframe_station(station, startdate=startdate, enddate=enddate):
    return pd.read_csv(
        stationdir / filepattern_station.format(ioac=stations[station]))
```

```python
wind_station = get_dataframe_station(station=teststation)
wind_station.index = pd.DatetimeIndex(wind_station.ts)
wind_station['month'] = wind_station.index.month.map(dict(enumerate(cal.month_abbr)))
wind_station
```

```python
ax = WindroseAxes.from_ax()
ax.bar(wind_onemonth.wdir10, wind_onemonth.wspd10, normed=True, 
    bins=[0.5, 6, 10, 14, 18, 22], nsector=32,
    opening=0.95, edgecolor="white", calm_limit=0.5,)
largesttick = ax.get_yticks()[-1]
largesttick = 2.5 * (1 + np.floor(largesttick / 2.5) )
ax.set_yticks(np.arange(2.5, largesttick, step=2.5))
ax.set_yticklabels([f"{percentage}%" for percentage in np.arange(2.5, largesttick, step=2.5)],
    horizontalalignment='left')

ax.set_legend()
ax.set_title(f"ERA5, 4 km, {locations[teststation]}, September winds")
```

```python

def plot_windrose_subplots(data, *, direction, var, color=None, **kwargs):
    """wrapper function to create subplots per axis"""
    ax = mpl.pyplot.gca()
    ax = WindroseAxes.from_ax(ax=ax)
    plot_windrose(direction_or_df=data[direction], var=data[var], ax=ax, **kwargs)
```

```python
g = sns.FacetGrid(
    data=wind_station,
    # the column name for each level a subplot should be created
    col="month",
    # place a maximum of 3 plots per row
    col_wrap=2,
    subplot_kws={"projection": "windrose"},
    sharex=False,
    sharey=False,
    despine=False,
    height=3.5,
)
g.map_dataframe(
    plot_windrose_subplots,
    direction="wd",
    var="ws",
    normed=True,
    # manually set bins, so they match for each subplot
    bins=[0.5, 6, 9, 12, 15, 18, 21], 
    nsector=16,
    opening=0.95, edgecolor="white",
    calm_limit=0.5,
    kind="bar",
)
for ax in g.axes:
    ax.set_rlabel_position(270)
g.fig.suptitle(f"{teststation} station")
```
