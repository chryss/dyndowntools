```python
from pathlib import Path
import xarray as xr
import rioxarray as rxr
import xwrf
```

```python
datadir1 = Path("/center1/DYNDOWN/cwaigl/ERA5_WRF/WPS202111_B_nc")
datadir2 = Path("/beegfs/CMIP6/wrf_era5/04km")
yr = "2021"
mth = "11"
day = "11"
hour = "00"
fp2 = "era5_wrf_dscale_4km"
fp1 = "met_em.d02."

testfp = datadir1 / f"{fp1}{yr}-{mth}-{day}_{hour}:00:00.nc" 
testfp2 = datadir2 / f"{yr}/{fp2}_{yr}-{mth}-{day}.nc"
```

```python
testds = xr.open_dataset(testfp, decode_coords='all', decode_times=True)
testds = testds.xwrf.postprocess()
```

```python
for var in testds.data_vars:
    testds[var].attrs['grid_mapping'] = 'wrf_projection'
```
