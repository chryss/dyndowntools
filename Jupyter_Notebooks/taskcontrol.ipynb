```python
import datetime as dt
import pandas as pd
```

```python
startstr = '221230'
startdate = dt.datetime.strptime(startstr, '%y%m%d').date()
startdate
```

```python
def get_bridge(date):
    if (
        date.day >= 26) or (
        date.day <= 10):
        return 1
    else:
        return 0
```

```python
records = []
thisdate = startdate

while True:
    ready = 0
    bridge = get_bridge(thisdate)
    records.append((thisdate.strftime('%y%m%d'), thisdate.year, thisdate.month, bridge, ready))
    thisdate = (thisdate - dt.timedelta(days=2))
    if thisdate.year < 1940:
        break

```

```python
recordsdf = pd.DataFrame.from_records(records, columns=['datelabel', 'year', 'month', 'bridgemonth', 'era5ready', ])
recordsdf['erapreproc'] = 0
recordsdf['wpsready'] = 0
recordsdf['wrfstarted'] = '' 
recordsdf['wrfcomplete'] = '' 
recordsdf['wrfsuccessful'] = 0
recordsdf['dataarchived'] = 0
recordsdf['wrfcleanup'] = 0
recordsdf['wpscleanup'] = 0
recordsdf['era5cleanp'] = 0
recordsdf 
```

```python
recordsdf.to_feather('status.feather')
```
