```python
import datetime as dt
import pandas as pd
```

```python
startstr = '221228'
startdate = dt.datetime.strptime(startstr, '%y%m%d').date()
startdate
```

```python
records = []
thisdate = startdate

while True:
    ready = 0
    if thisdate.month in [11, 12]:
        ready = 1
    bridge = 0
    if (
        thisdate.month != (thisdate + dt.timedelta(days=2)).month
        ) or (
        thisdate.month != (thisdate - dt.timedelta(days=1)).month):
        bridge = 1
    records.append((thisdate.strftime('%y%m%d'), ready, bridge))
    thisdate = (thisdate - dt.timedelta(days=2))
    if thisdate.year < 2020:
        break

```

```python
recordsdf = pd.DataFrame.from_records(records, columns=['datelabel', 'inputready', 'bridgemonth'])
recordsdf['wpssetup'] = '' 
recordsdf['wrfsetup'] = '' 
recordsdf['wrfprocess'] = '' 
recordsdf['filearchive'] = '' 
recordsdf['filepostproc'] = '' 
recordsdf['cleanup'] = ''
recordsdf 
```

```python
recordsdf.to_feather('status.feather')
```
